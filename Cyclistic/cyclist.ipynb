Mayoritas kolom pada dataset ini memiliki tipe data ***object*** (*string or mixed types*). Data frame ini tercatat tidak memiliki nilai *null* , maka dari itu kita tidak perlu repot-repot untuk membersihkan nilai *null*.

Hanya 4 kolom yang memiliki tipe data ***float64***. Agar memudahkan proses analisis, kita akan mengubah tipe data pada kolom ***started_at*** dan ***ended_at*** menjadi tipe data datetime

## Kolom Numerik
#### Ride_Length

Kolom ini berisikan informasi mengenai berapa lama tiap pengguna menggunakan sepeda Cyclistic. Untuk mempermudah proses analisis, kita akan mengubahnya ke dalam satuan menit dengan membuat kolom baru bernama `duration` 

```python
# Mengubah satuan waktu menjadi menit
df['duration'] = df.ride_length / np.timedelta64(1,"m")

# Menampilkan informasi statistik kolom duration
df.duration.describe().apply("{0:.5f}".format)
```

Dari data di atas, kita dapat melihat rata-rata lama pengguna menggunakan sepeda yaitu **21,94 menit**. Lalu untuk penggunaan paling lama itu ada di angka **55944,15 menit** atau setara dengan **38 hari 20 jam 24 menit**.
Sebelum membuat visualisasi data dari kolom `ride_length`, kita akan melihat apakah ada nilai pencilan (*outlier*) pada data frame. Untuk mengetahuinya kita akan menggunakan metode *IQR (Inter Quartile Range)*

```python
# IQR
Q1 = np.percentile(df.duration, 25,
                   interpolation = 'midpoint')
 
Q3 = np.percentile(df.duration, 75,
                   interpolation = 'midpoint')
IQR = Q3 - Q1
# IQR
print("Nilai dari IQR adalah "+str(IQR))
```

```python

# Above Upper bound
upper = (Q3+1.5*IQR)
 
print("Batas Atas:",upper)
# print(np.where(upper))
 
# Below Lower bound
lower = (Q1-1.5*IQR)
print("Batas Bawah:", lower)
# print(np.where(lower))


# # Above Upper bound
# upper = df.duration >= (Q3+1.5*IQR)
 
# print("Upper bound:",upper.sum())
# # print(np.where(upper))
 
# # Below Lower bound
# lower = df.duration <= (Q1-1.5*IQR)
# print("Lower bound:", lower.sum())
# # print(np.where(lower))
```

```python
# Menghitung jumlah data yang dikategorikan sebagai outlier berdasarkan metode IQR
outlier = df.duration[(df.duration >= upper) | (df.duration <= lower)].shape[0]
print("Jumlah outlier = "+ str(outlier))

# Perbandingan outlier dengan total data\
outlier_pct = (outlier / df.shape[0])*100
print("Sebanyak "+"{:.2f}".format(outlier_pct)+"% dari data yang ada pada data frame merupakan outlier")
```

Sebenarnya ada beberapa metode untuk mengatasi data outlier ini, namun untuk saat ini kita akan mengabaikannya dan tetap melanjutkan dengan data yang ada.

Selanjutnya kita lihat rata-rata ride_length / durasi untuk setiap day_of_week dan distribusi ride_length menggunakan bar chart dan juga boxplot

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(16,8)) # Membuat Subplot
# ax1.hist(df.duration, bins=np.arange(2,150,2)) # Membuat histogram
(df.duration.groupby(df.day_of_week).mean()).plot(kind='bar',ax=ax1)
ax1.set_ylabel("Menit")
ax1.set_xlabel("Day_of_Week")
ax1.set_title('Rata-Rata Durasi Pengguna (Menit)')
ax2.boxplot(df['duration']) # Membuat boxplot
ax2.set_ylabel('Durasi Penggunaan')
ax2.set_title('Box Plot')
plt.suptitle("Pengguna Berdasarkan Durasi Penggunaan")
plt.show() 
```

Dilihat dari distribusinya, ternyata data frame memiliki distribusi tidak normal / condong ke kanan. Hal ini bisa kita lihat dari histogram yang ekor sebelah kanannya lebih panjang (mempunyai kecondongan positif)

### Bagaimana anggota tahunan dan pengendara biasa menggunakan sepeda sewaan secara berbeda?

Untuk menjawab pertanyaan ini, kita perlu mengubah kolom `member_casual` dari kategori menjadi numerik agar mempermudah analisis. Selain itu, kita juga perlu membuat syntax untuk memudahkan pembuatan visualisasi

```python
# Untuk mengubah kolom member_casual dari kategorikal menjadi numerikal
membership = pd.get_dummies(df, columns=['member_casual'])

# Membuat data frame khusus untuk pengguna casual

start_top10_casual = df[['start_station_name','member_casual']][(df.start_station_name.isin(top10_start_station_name.index))&(df.member_casual == 'casual' )].groupby('start_station_name').count()

# Membuat data frame khusus untuk member

start_top10_member = df[['start_station_name','member_casual']][(df.start_station_name.isin(top10_start_station_name.index))&(df.member_casual == 'member' )].groupby('start_station_name').count()

# # Melihat Top 10 stasiun awal berdasarkan tipe keanggotaan
# df[['ride_id','start_station_name','member_casual']][df.start_station_name.isin(top10_start_station_name.index)].groupby(['start_station_name','member_casual']).count()

# Membuat data frame khusus untuk pengguna casual

end_top10_casual = df[['end_station_name','member_casual']][(df.end_station_name.isin(top10_end_station_name.index))&(df.member_casual == 'casual' )].groupby('end_station_name').count()

# Membuat data frame khusus untuk member

end_top10_member = df[['end_station_name','member_casual']][(df.end_station_name.isin(top10_end_station_name.index))&(df.member_casual == 'member' )].groupby('end_station_name').count()

# # Melihat Top 10 stasiun awal berdasarkan tipe keanggotaan
# df[['ride_id','end_station_name','member_casual']][df.end_station_name.isin(top10_end_station_name.index)].groupby(['end_station_name','member_casual']).count()

member = df[df.member_casual == 'member']
casual = df[df.member_casual == 'casual']
```

```python
fig, axes = plt.subplots(3,3, figsize=(16,24))

# Visualisasi Jumlah pelanggan berdasarkan tipe sepeda
sns.countplot(data=df, x='rideable_type', order=df.rideable_type.value_counts().index, hue='member_casual', ax=axes[0,0])
axes[0,0].set_xlabel('Tipe Sepeda')
axes[0,0].set_ylabel('Jumlah Pengguna')
axes[0,0].set_title('Pengguna Berdasarkan Tipe Sepeda')

# Visualiasi
x = np.arange(10)
width = 0.4

axes[0,1].bar(x-0.2,start_top10_member.member_casual, width, label='Member')
axes[0,1].bar(x+0.2,start_top10_casual.member_casual, width, label='Casual')
axes[0,1].set_xticks(x)
axes[0,1].set_xticklabels(start_top10_member.index, rotation=90)
axes[0,1].legend(['Member','Casual'])
axes[0,1].set_ylabel('Jumlah Pengguna')
axes[0,1].set_title('Pengguna Berdasarkan Top 10 Stasiun Awal')

# Visualisasi 3
x = np.arange(10)
width = 0.4

axes[0,2].bar(x-0.2,end_top10_member.member_casual, width, label='Member')
axes[0,2].bar(x+0.2,end_top10_casual.member_casual, width, label='Casual')
axes[0,2].set_xticks(x)
axes[0,2].set_xticklabels(end_top10_member.index, rotation=90)
axes[0,2].set_ylabel('Jumlah Pengguna')
axes[0,2].legend(['Member','Casual'])
axes[0,2].set_title('Pengguna Berdasarkan Top 10 Stasiun Akhir')


# Visualisasi 4
z = np.arange(7)
width = 0.4
axes[1,0].bar(z-0.2,member.duration.groupby(member.day_of_week).mean(), width, label='member')
axes[1,0].bar(z+0.2,casual.duration.groupby(casual.day_of_week).mean(), width, label='casual')
axes[1,0].legend(['Member','Casual'])
axes[1,0].set_xlabel("Day_of_Week")
axes[1,0].set_ylabel("Menit")
axes[1,0].set_title('Rata-Rata Durasi Berdasarkan day_of_week')

# Visualisasi 5
axes[1,1].bar(z-0.2,membership['member_casual_member'].groupby(membership.day_of_week).sum(), width, label='member')
axes[1,1].bar(z+0.2,membership['member_casual_casual'].groupby(membership.day_of_week).sum(), width, label='casual')
axes[1,1].set_xticks(z)
axes[1,1].set_xticklabels(z+1, rotation=0)
axes[1,1].legend(['member','casual'])
axes[1,1].set_xlabel('Day_of_Week\n(1= Minggu 7= Sabtu)')
axes[1,1].set_ylabel('Jumlah Pengguna')
axes[1,1].set_title('Jumlah Pengguna Berdasarkan day_of_week')

# Visualisasi 6
axes[1,2].plot(membership['member_casual_member'].groupby(membership.started_at.dt.month).sum())
axes[1,2].plot(membership['member_casual_casual'].groupby(membership.started_at.dt.month).sum())
axes[1,2].legend(['member','casual'])
axes[1,2].set_xlabel('Bulan')
axes[1,2].set_ylabel('Jumlah Pengguna')
axes[1,2].set_title('Pengguna Bulanan (Waktu Mulai)')

# Visualisasi 7
axes[2,0].plot(membership['member_casual_member'].groupby(membership.started_at.dt.hour).sum())
axes[2,0].plot(membership['member_casual_casual'].groupby(membership.started_at.dt.hour).sum())
axes[2,0].legend(['member','casual'])
axes[2,0].set_xlabel('Jam')
axes[2,0].set_ylabel('Jumlah Pengguna')
axes[2,0].set_title('Pengguna Berdasarkan Jam (Waktu Mulai)')

#Visualisasi 8
axes[2,1].plot(membership['member_casual_member'].groupby(membership.ended_at.dt.month).sum())
axes[2,1].plot(membership['member_casual_casual'].groupby(membership.ended_at.dt.month).sum())
axes[2,1].legend(['member','casual'])
axes[2,1].set_xlabel('Bulan')
axes[2,1].set_ylabel('Jumlah Pengguna')
axes[2,1].set_title('Pengguna Bulanan (Waktu Berakhir)')

# Visualisasi 9
axes[2,2].plot(membership['member_casual_member'].groupby(membership.ended_at.dt.hour).sum())
axes[2,2].plot(membership['member_casual_casual'].groupby(membership.ended_at.dt.hour).sum())
axes[2,2].legend(['member','casual'])
axes[2,2].set_xlabel('Jam')
axes[2,2].set_ylabel('Jumlah Pengguna')
axes[2,2].set_title('Pengguna Berdasarkan Jam (Waktu Berakhir)')

plt.suptitle('Member VS Casual', fontweight='bold', fontsize=18)
plt.tight_layout(pad=2);
plt.show()

```

### Temuan Penting

- `docked_bike` didominasi oleh **pengguna harian / casual**, berbeda dengan 2 tipe sepeda lainnya yang didominasi oleh **pengguna tahunan / member**
- Rata-rata penggunaan sepeda **pengguna harian / casual** 2 kali lebih tinggi dibandingkan **pengguna tahunan / member** 
- Jumlah **pengguna tahunan / member** cenderung stabil setiap harinya namun jumlah penggunanya cenderung lebih banyak pada saat weekdays (Senin-Jumat). Sedangkan **pengguna harian / casual** justru lebih banyak jumlahnya pada saat weekend (Sabtu-Jumat) tiba dan cenderung menurun pada saat weekdays. Dari informasi itu, kita dapat membuat dugaan sementara / hipotesis bahwa **pengguna tahunan / member** kebanyakan anggotanya menjadikan Cyclistic sebagai moda transportasi sehari-hari sedangkan **pengguna harian / casual** menjadikan Cyclistic hanya sebagai untuk bersantai di akhir pekan.
- Kedua tipe keanggotaan memiliki tren bulanan yang cukup mirip dan keduanya mengalami peningkatan di sekitar bulan Juni sampai September. Kemungkinan peningkatan ini terjadi karena waktunya yang bersamaan dengan periode liburan musim panas ataupun cuaca yang cenderung lebih hangat dibanding bulan-bulan lain
